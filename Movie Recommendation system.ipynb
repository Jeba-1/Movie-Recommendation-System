# genres,keywords,overview,title,cast,crew,movie_id

```python
movies = movies[['movie_id','title','genres','keywords','overview','cast','crew']]
movies.head()
```

```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter <3:
            L.append(i['name'])
        counter+=1
    return L
            
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new=movies.drop(columns=['overview','genres','cast','crew'])
new.head()
```

```python
new.info()
```

```python
new['tags']=new['tags'].apply(lambda x: " ".join(x))
```

```python
new.head()
```

```python
new.to_csv('new_dataset.csv')
```

# NLP concept-Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',max_features=5000,binary=True)
```

```python
vector=cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

# Calculate Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity matrix
similarity = cosine_similarity(vector)

# Display the shape of the similarity matrix
print(similarity.shape)
```

# Function to get Movie Recommendations

```python
def recommend(movie):
    if movie not in new['title'].values:
        print(f"Movie '{movie}' not found in the dataset.")
        return
    
    movie_index = new[new['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    print(f"Recommendations for '{movie}':")
    for i in movies_list:
        print(new.iloc[i[0]].title)

# Test the recommendation system
recommend('Avatar')
print('---------------------------------------------------------')
recommend('The Dark Knight')
print('---------------------------------------------------------')
recommend('Inception')
print('---------------------------------------------------------')
recommend('The Avengers')
print('---------------------------------------------------------')
recommend('Interstellar')
print('---------------------------------------------------------')
recommend('Titanic')
```
